```python
from csdmpy.data.ssda903 import SSDA903TableType
from csdmpy.datacontainer import DemandModellingDataContainer
import csdmpy_fakedata as sample
```

# Datacontainer API

At the core of the datafiles api is the ability to read, index and merge the data files. 

To initialse the system, pass in a DataStore

```python
dc = DemandModellingDataContainer(sample.V1)

dc.first_year, dc.last_year
```

We can then access the files directly using the year and table type

```python
dc.get_table(dc.first_year, SSDA903TableType.EPISODES)
```

## Combining the files

A key part of the demand modelling is a unified, combined view of several years. To do this the Headers and Episodes needs to be combined to
create a child-based view. This view can be fetched on a year-by-year basis.

```python
dc.get_combined_year(dc.first_year)
```

Next these files are combined and de-duplicated so we end up with a single, multi-year view.

```python
dc.get_combined_data()
```

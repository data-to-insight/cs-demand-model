```python
df_comp = pd.concat([df_o.end_age, df_n.end_age], axis=1)
df_comp.columns=['new', 'old']
df_comp['age_diff'] = np.abs(df_comp.new - df_comp.old)
df_comp = df_comp.query('age_diff > 1')
df_comp
```
